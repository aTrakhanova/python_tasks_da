### Сколько вакансий, которые вам нравятся?

```python
from collections import Counter

# Вакансии d-people

data_vacancies = [row for row in vacs if row['vactitle'].count('Data')]
len(data_vacancies)
```

```python
{row['vactitle'] for row in data_vacancies}
```

### Насколько свежие эти вакансии?

```python
from collections import defaultdict

vacancy_date = defaultdict(list)
for row in data_vacancies:
    vacancy_date[row['vacdate']].append(row)
```

```python
vacancy_date.keys()
```

```python
vacancy_date['2020-10-23']
```

### Сколько вакансий для аналитика данных?

```python
DA_position = [row for row in vacs_copy if row['vactitle'].count('DATA ANALYST')]
len(DA_position)
```
